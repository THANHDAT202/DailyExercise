```python
#load library
import pandas as pd
#create dataframe
df = pd.DataFrame()
#add columns
df['Name'] = ['Lemon','Apple','Banana']
df['price'] = [5,8,7]
df['License'] = [False, True, True]
#show data
df
```

```python
#hàm .loc giúp bạn lấy ra được các hàng tương ứng 
#.loc[dieu kien,[ten_cac_cot]]
#df.loc[df.quatity>5,['q]]
df.loc[df.price > 5]
```

```python
df.loc[df.price == 5,['Name']]
```

```python
#hàm iloc
#lấy ra các hàng trùng với số thứ tự, có thể chỉ định các cột trả về tương tự với hàm loc
#cáu trúc: .iloc[từ hàng 1 tới hàng 5,từ cột A tới cột C] == .iloc[1:5,A:C]
df.iloc[0:2]
```
